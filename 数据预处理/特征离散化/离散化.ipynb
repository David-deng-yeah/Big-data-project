```python
Block32 = pd.read_csv('../../RSP数据块/Block 32.csv')
Block42 = pd.read_csv('../../RSP数据块/Block 42.csv')

X_rsp32 = Block32.drop('label',axis=1)
y_rsp32 = Block32['label']
print('X_rsp32 shape: ',X_rsp32.shape)

X_rsp42 = Block42.drop('label',axis=1)
y_rsp42 = Block42['label']
print('X_rsp42 shape: ',X_rsp42.shape)
```

```python
X_rsp32.head()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
fea = X_rsp32.columns
X_tmp = deepcopy(X_rsp32)
# 分箱为10份
Xt = KBinsDiscretizer(n_bins=20, encode='ordinal', strategy='uniform').fit_transform(X_tmp)
df = pd.DataFrame(data=Xt, columns=fea)
df
```

```python
plt.style.use('seaborn')

fig, axes = plt.subplots(ncols=7, nrows=4,figsize=[50,30])
for i, ax in zip(range(0,28), axes.flat):
    ax.set_title('fea'+str(i))
    ax.set_xlabel('values')
    ax.set_ylabel('nums')
    sns.distplot(df.iloc[:,i],
        hist=True,kde=False, ax=ax,color='royalblue')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier().fit(Xt,y_rsp32.values)
```
